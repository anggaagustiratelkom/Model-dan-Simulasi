```
import matplotlib.pyplot as plt
import math

x, y, x1, y1, t, vy, vx = [], [], [], [], [], [], []

x.append(0)
y.append(0)
x1.append(0)
y1.append(0)
v0 = 50
deg = 35
g = 9.806
pi = 22/7
theta = deg*pi/180
vx = v0*math.cos(theta)
vy.append(v0*math.sin(theta))
ax = 0
ay = -g
T = ((-2*v0*math.sin(theta))/(ay))
dt = 0.01
N = int((T/dt)+1)

t.append(0)
for i in range(N):
  t.append(t[i]+dt)

j = 1
while j<= N:
  x1.append((0.5*ax*(t[j]*t[j]))+(vx*t[j]))
  y1.append((0.5*ay*(t[j]*t[j]))+(vy[0]*t[j]))
  j += 1

i = 1
while i<= N:
 vy.append(vy[i-1]+(ay*dt))
 y.append(y[i-1]+(vy[i]*dt))
 x.append(x[i-1]+(vx*dt))
 i += 1

plt.plot(x1, y1) #blue, for numerik
plt.plot(x, y) #orange, for analitik
plt.show()
print(vy)
print(vx)
print(ax)
print(ay)
# print(theta)
# print(math.sin(theta))
print(T)
print(N)
```
